### Examining the Categorical variables

```python
df.columns
```

```python
df_cat = df[['title_simple','Size','Industry','company_name','Location','Headquarters','Type of ownership','Sector','Revenue','python', 'spark', 'aws', 'excel', 'tableau','sql', 'seniority']]
```

```python
for cat in df_cat.columns:
    print("Graph of {}; Total Categories: {}".format(cat,len(df[cat].unique())))
    if len(df[cat].unique())>20:
        plot = sns.barplot(x=df[cat].value_counts()[:20].index,y=df[cat].value_counts()[:20])
    else:
        plot = sns.barplot(x=df[cat].unique(),y=df[cat].value_counts())
    plot.set_xticklabels(plot.get_xticklabels(), rotation =90)
    plt.show()
```

### Analyze average salary in relation to different variables using Pivot tables

```python
df.columns
```

```python
df_pivots = df[['Rating','state','Location','Headquarters','Size','Type of ownership','Industry','Sector','Revenue','title_simple']]
```

```python
for column in df_pivots.columns:
    print('\n',column)
    print(df.pivot_table(index=column, values='avg_salary').sort_values('avg_salary', ascending=False))
```
